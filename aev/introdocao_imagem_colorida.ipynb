# Imagens Coloridas

Uma imagem em tons de cinza pode ser representada por uma matriz, onde cada elemento da matriz (valores de 0 a 255) representa um tom de cinza da imagem (de preto a branco, respectivamente).

Uma imagem coloriada Ã© representada por uma matriz tridimensional (vertical x horizontal x cor). Usualmente adota-se o sistema RBG (vermelho, verde, azul). Vejamos um exemplo: 

```octave
I = imread('imgs/lena.png');
Id = im2double(I); % conversao utilizada para contornar um problema no kernel Octave para Jupyter notebook
imshow(Id);
whos
```

Note que a image colorida Ã© presentada por uma matriz 256x256x3 (256x256 pixels em RGB). Podemos separar as componentes RBG e visualizÃ¡-las separadamente. Iremos utilizar um colormap vermelho, verde e azul para cada uma das componentes.

```octave
figure("position",get(0,"screensize")./[1 1 2 3]);
subplot(1,3,1); cmap = [[0:255]'/255, zeros(256,2)]; imshow(I(:,:,1),cmap); title('R'); 
subplot(1,3,2); cmap = [zeros(256,1), [0:255]'/255, zeros(256,1)];  imshow(I(:,:,2),cmap); title('G');
subplot(1,3,3); cmap = [zeros(256,2), [0:255]'/255]; imshow(I(:,:,3),cmap); title('B');
```

Caso queiramos trabalhar com a imagem em tons de cinza, podemos converter utilizando a funÃ§Ã£o *rgb2gray*. O resultado serÃ¡ uma matriz 256x256 representando os tons de cinza da imagem.

```octave
Ipb = rgb2gray(I);
imshow(Ipb);
whos
```

Vamos agora ler uma imagem GIF que foi criada usando uma palheta de cores (palheta web 256, removidas as cores que nÃ£o foram utilizadas na imagem). Note que agora a imagem *I* serÃ¡ uma matriz 256x256, onde o valor de cada elemento da matriz corresponde a um Ã­ndice que referencia uma cor na palheta de cores, armazenada na variÃ¡vel *map*.

```octave
clear all;
[I,map] = imread ('imgs/lena.gif');
whos
```

Para exibir a imagem, devemos utilizar a palheta de cores utilizada para codifica tal imagem.

```octave
imshow(I,map);
```

O valores da matriz *I* variam de 0 a 54, representando os Ã­ndices para as cores na palheta.

```octave
unique (I(:))'
```

Note que a palheta possui 55 cores. A matriz *map* possui 64 linhas apenas pelo fato de terem sido utilizados 6 bits para indexar cada pixel da imagem. Podemos disprezar as linhas 56 a 64. 

```octave
map(55:end,:)
map = map(1:55,:);
```

```octave
figure("position",get(0,"screensize")./[1 1 2 3]);
subplot(1,2,1); imshow(I,map); 
subplot(1,2,2); imshow([0:55]',map); title('palheta de cores utilizada');
```
